```julia
using Plots
using LaTeXStrings
using DelimitedFiles
```

```julia
timming_file2 = raw"./12_whit_openblas_04schedule_chunksize/result_collection_total_timing_data.dat";
timming_matrix2 = readdlm(timming_file2);
chunksize=["$(timming_matrix2[i+1,1])" for i in 1:length(timming_matrix2[2:end,1])];
plot(chunksize[1:end],timming_matrix2[2:end,2],seriestype =:bar,color="red",label="")
plot!(title="Resultados \"speed file\", \$λ=0.2\$ \n pthread=2, OpenBLAS, (dynamic,chunksize)",xlabel="Chunksize",ylabel="Real time[s]")
```
